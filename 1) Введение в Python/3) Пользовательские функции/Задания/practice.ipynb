```python
def read_tasks(file_name):
    with open(file_name, 'r', encoding='utf-8') as f:
        tasks_list = list(map(lambda x: x[:-1] if x[-1]=='\n' else x, f.readlines())) 
        return tasks_list

read_tasks('tasks.txt')        
```

```python
def read_tasks(text_file):
    tasks_list = []
    with open(file=text_file, mode='r', encoding='utf-8') as f:
        for line in f:
            strip_line = line.strip()
            tasks_list.append(strip_line)
    return tasks_list
    
read_tasks('tasks.txt')     
```

```python
def calculate_cr(purchases, visits):
    if visits != 0:
        onversion_rate = purchases*100/visits
        print(f'  : {round(onversion_rate, 2)}%')
    else:
        print(f'   0,   .')    
```

```python
def calculate_cr_upd(list_purchases, list_visits):
    list_onversions = []
    for i in range(len(list_visits)):
        if list_visits[i] != 0:
            onversion = round(list_purchases[i]*100/list_visits[i], 2)
            list_onversions.append(onversion)
        else:
            list_onversions.append(None)
    return(list_onversions)
```

```python
            
purchases_list = [10, 5, 8, 0, 0, 12]
visits_list = [500, 650, 730, 0, 80, 320]
calculate_cr_upd(purchases_list, visits_list)       
        
```

```python
def count_months(data, threshold_value):
    counter = 0
    for cortej in data:
        if cortej[1]>threshold_value:
            counter += 1
    print(f'     {threshold_value}: {counter}')
```

```python
def filter_months(data, threshold_value):
    months = []
    for cortej in data:
        if cortej[1]>threshold_value:
            months.append(cortej[0])
    return(months)
```

```python
def stream_gen(name, count):
    for n in range(1, count+1):
        yield(f'{name}_{n}')
```

```python
def apply_discount(dict, sale_rate=0):
    new_dict = {}
    for name in dict.keys():
        new_price = dict[name]*(100-sale_rate)//100
        new_dict[name] = new_price
    return new_dict
    
```

```python
dict = {
    'StartDA': 80000, 
    'StartML': 150000, 
    'StartDE': 90000, 
    'HardDA': 120000, 
    'HardML': 150000
}
sale_rate = 10
apply_discount(dict, sale_rate)
```

```python
def calculate_metric(purchases, metric = 'revenue'):
    if metric not in ('revenue', 'orders', 'paying_users', 'aov', 'aov_alt'):
        print('  .')
    else:
        if metric == 'revenue':
            revenue = 0
            for user_buys in purchases:
                revenue += sum(user_buys)
            return(revenue)

        
        elif metric == 'orders':
            counter = 0
            for user_buys in purchases:
                counter += len(user_buys)
            return(counter)
        
        elif metric == 'paying_users':
            return(len(purchases))
            
        elif metric == 'aov':
            return(round(calculate_metric(purchases, 'revenue')/calculate_metric(purchases, 'orders'), 2))

        elif metric == 'aov_alt':
            revenues = []
            for user_buys in purchases:
                revenues.append(sum(user_buys)/len(user_buys))
            return(round(sum(revenues)/len(revenues), 2))
```

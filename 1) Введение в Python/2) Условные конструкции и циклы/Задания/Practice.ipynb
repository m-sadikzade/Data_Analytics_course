```python
worker_experience = [
    ['Иван', 'Иванов', 2],
    ['Анатолий', 'Карпов', 10],
    ['Кирилл', 'Кириллов', 4],
    ['Пётр', 'Петров', 3],
    ['Константин', 'Константинов', 8],
    ['Андрей', 'Андреев', 3],
    ['Ольга', 'Петрова', 3],
    ['Ирина', 'Иришкова', 3],
    ['Фёдор', 'Фёдоров', 5],
    ['Семён', 'Семёнов', 0]
]

worker_department = {
    'Иван Иванов': 'Отдел маркетинга',
    'Семён Семёнов': 'Отдел аналитики',
    'Анатолий Карпов': 'Отдел аналитики',
    'Кирилл Кириллов': 'Отдел аналитики',
    'Пётр Петров': 'Отдел продаж',
    'Ирина Иришкова': 'Отдел продаж',
    'Фёдор Фёдоров': 'Отдел аналитики',
    'Ольга Петрова': 'Отдел аналитики',
    'Константин Константинов': 'Отдел разработки',
    'Андрей Андреев': 'Отдел маркетинга'
}
```

```python
analyst_grades = {}
for worker in worker_experience:
    name = f'{worker[0]} {worker[1]}'
    department = worker_department[name]

    if department == 'Отдел аналитики':
        experience = worker[2]
        if experience < 2:
            level = 'junior'
        elif 2 <= experience <= 5:
            level = 'middle'
        else:
            level = 'senior'
            
        analyst_grades[name] = {'experience': experience, 'grade': level}  
```

```python
student_data = {
    'alice.dem': {'name': 'Алиса', 'surname': 'Демидова', 'points': [5, 5, 5, 5, 5, 10]},
    'mikhail.val': {'name': 'Михаил', 'surname': 'Валетов', 'points': [3, 3, 5, 1, 4, 8]},
    'anton.kar': {'name': 'Антон', 'surname': 'Каргин', 'points': [5, 1, 5, 2, 0, 5]},
    'maria.vol': {'name': 'Мария', 'surname': 'Волкова', 'points': [4, 4, 5, 4, 3, 7]},
    'roman.tok': {'name': 'Роман', 'surname': 'Токарев', 'points': [2, 3, 5, 3, 3, 0]},
}
```

```python
student_results = {}
for student in student_data.values():
    name = student['name'] + ' ' + student['surname']
    points_avg = sum(student['points'][:5])/5
    final_point = student['points'][-1]
    check = (points_avg) > 3 and (final_point >= 5)
    result = 'failed'
    if check == True:
        result = 'passed'  
    student_results[name] = {'points_avg': points_avg, 'points_final': final_point, 'result': result}
```

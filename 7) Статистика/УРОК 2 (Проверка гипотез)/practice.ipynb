```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
res = data.groupby('fb_campaign_id').agg({'Impressions':'sum'})
np.log(res)
```

```python
plt.figure(figsize=(30, 8))
sns.barplot(data=np.log(res).reset_index(), x='fb_campaign_id', y='Impressions')
```

```python
camaigns = data.xyz_campaign_id.unique()
```

```python
for i, cp in enumerate(camaigns):
    res = data.query('xyz_campaign_id == @cp')
    plt.figure(i)
    ax = sns.distplot(res['CTR'], bins=20)
    ax.set_title(f'xyz_campaign_id = {cp}')
```

```python
from scipy import special #модуль питона, предназначенный для научных и инженерных расчетов
```

```python
p = 0.4 # вероятность отказа элемента устройства
n = 5   # количество элементов устройства
k = 3   # откажут 3 элемента

#Вероятность отказа 3 элесентов из 5
a = p*p*p*(1-p)*(1-p)

#Количество комбинаций
b = special.comb(n, k)

res = a*b
round(res, 4)
```

```python
p = 0.4 # вероятность отказа элемента устройства
n = 5   # количество элементов устройства
k = 1   # откажет 1 элемент

# ни один элемент не откажет 
a = (1-p)*(1-p)*(1-p)*(1-p)*(1-p)

#откажет хотя бы один
res = 1-a
round(res, 3)

```

SELECT count(1) as persons_cnt,
       max(coalesce(year_of_death - year_of_birth, date_part('year', now())::int - year_of_birth)) as max_age,
       min(coalesce(year_of_death - year_of_birth, date_part('year', now())::int - year_of_birth)) as min_age,
       avg(coalesce(year_of_death - year_of_birth, date_part('year', now())::int - year_of_birth)) as avg_age
FROM   persons;
